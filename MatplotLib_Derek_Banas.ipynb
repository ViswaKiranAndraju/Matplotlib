```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_1 = np.linspace(0,5,10)
y_1 = x_1**2
plt.title("Squared chart")
plt.xlabel("x")
plt.ylabel("x^2")
plt.plot(x_1,y_1)
```

```python
plt.subplot(1,2,1)
plt.plot(x_1,y_1,'r')
plt.subplot(1,2,2)
plt.plot(x_1,y_1,'b')
```

```python
fig_1 = plt.figure(figsize=(6,4),dpi=100)
axes_1 = fig_1.add_axes([0.1,0.1,0.9,0.9])
axes_1.set_xlabel("x")
axes_1.set_ylabel("x^2")
axes_1.set_title("Sqaured chart")
axes_1.plot(x_1,y_1,'r',label = "x^2/x")
axes_1.plot(y_1,x_1,'k',label = "x/x^2")
axes_1.legend()

axes_2 = fig_1.add_axes([0.4,0.4,0.4,0.5])
axes_2.set_xlabel("x")
axes_2.set_ylabel("x^2")
axes_2.set_title("Sqaured")
axes_2.plot(x_1,y_1,label = "x/x^2")
axes_2.plot(y_1,x_1,label = "x^2/x")
axes_2.legend()

axes_1.text(0,22,"Hello")
```

```python
fig_2, axes_2 = plt.subplots(figsize=(6,4),nrows=1,ncols=3)
plt.tight_layout()
axes_2[1].set_title('plot 2')
axes_2[1].set_xlabel('x axis')
axes_2[1].set_ylabel('y axis')
axes_2[1].plot(x_1,y_1)
```

```python
ice_df = pd.read_csv('ice.csv')
ice_df = ice_df.sort_values(by="Temperature")

np_arr = ice_df.values
x_2 = np_arr[:,0]
y_2 = np_arr[:,1]
fig_4 = plt.figure(figsize=(6,4))
axes_4 = fig_4.add_axes([0.1,0.1,0.9,0.9])
axes_4.set_title("Ice creams vs temperature")
axes_4.set_xlabel("Temperature")
axes_4.set_ylabel("Ice creams")
axes_4.plot(x_2,y_2)

axes_4.annotate("Good month",xy = (83,536), xytext=(70,520), arrowprops=dict(facecolor='k',shrink = 0.05))
```
