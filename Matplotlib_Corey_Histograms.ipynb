```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
ages = [18, 19, 21, 25, 26, 26, 30, 32, 38, 45, 55]

plt.legend()

plt.hist(ages,bins=5,edgecolor='black')

plt.title('Ages by Total')
plt.xlabel('Ages')
plt.ylabel('Total')

plt.tight_layout()
```
